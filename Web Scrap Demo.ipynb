```python
#Importing Necessary Packages for Website Scraping 
import bs4 as bs
import urllib3 as urllib
import urllib.request
```

```python
#Setting timestamp to tag project for documentation purposes
import datetime
now=datetime.datetime.now()
```

```python
#Using scraping package to read in URL
testurl = 'https://www.tripadvisor.com/Restaurant_Review-g60745-d1937941-Reviews-Island_Creek_Oyster_Bar-Boston_Massachusetts.html'
testpage = urllib.request.urlopen(testurl)
soup = bs.BeautifulSoup(testpage,'html.parser')
```

```python
#(DEBUGGING CODE)Printing retrieved HTML code for website
soup
```

```python
#(DEBUGGING CODE)
soup.findAll('p',{'class':"partial_entry"})
```

```python
#"a" represents all the hyperlink in the webpage (Debugging code)
soup.find_all('a')
```

```python
#looping through to display all review entries on page
q = 1
for reviews in soup.findAll('p',{'class':"partial_entry"}):
    print(q)
    q += 1 
    print (reviews.text)
```

```python
#Creating a function to read in URL automatically 
def read_web(url):
    page = urllib.request.urlopen(url)
    result = bs.BeautifulSoup(page,'html.parser')
    return result
```

```python
#creating a function to print out reviews on page
def print_web(soup):
    print(now)
    q = 1
    for reviews in soup.findAll('p',{'class':"partial_entry"}):
        print(q)
        q += 1 
        print (reviews.text)
```

```python
source = read_web('https://www.tripadvisor.com/Restaurant_Review-g60745-d1937941-Reviews-Island_Creek_Oyster_Bar-Boston_Massachusetts.html')
```

```python
#Testing Result of source
source
```

```python
#Printing Reviews
print_web(source)
```
